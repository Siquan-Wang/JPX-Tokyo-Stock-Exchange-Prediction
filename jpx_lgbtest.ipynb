```python
import os
from pathlib import Path

import base64
```

```python
jpx_lgb = ''
```

```python
R = 5

for i in range(R):
    jpx_lgb = base64.b64decode(jpx_lgb)
```

```python
path = Path('./jpx_lgb.py')
path.parent.mkdir(exist_ok=True)
path.write_bytes(jpx_lgb)
```
